```python
from pytube import YouTube, Playlist
from youtube_transcript_api import YouTubeTranscriptApi
import json
```

```python
!mkdir ytd
!pwd
```

```python


out_dir = "/content/ytd"

def download_ytvideo_and_sub(yturl, out_dir=out_dir):
  
  yt = YouTube(yturl)

  yt.streams.filter(file_extension="mp4")

  title = yt.title.replace(" ","")
  print("start download {}".format(title))

  yt.streams.get_by_itag(22).download(output_path=out_dir,filename=title)


  transcript_list = YouTubeTranscriptApi.list_transcripts(yt.video_id)
  transcript_list

  transcript = transcript_list.find_transcript(['en'])
  translated_transcript = transcript.translate('zh-Hans')
  zh_trans = translated_transcript.fetch()
  en_trans = transcript.fetch()
  fname = out_dir+"/en_{}.json".format(title)
  with open(fname, "w") as fh:
      json.dump(en_trans, fh)
  fname = out_dir+"/zh_{}.json".format(title)
  with open(fname, "w") as fh:
      json.dump(zh_trans, fh)
      
  fname =  out_dir+"/en_zh_{}.json".format(yt.title.replace(" ",""))
  with open(fname, "w") as fh:
      json.dump(zh_trans+en_trans, fh)
  print("finish download {}".format(title))
```

```python
yturl = "https://www.youtube.com/watch?v=nhUKHf-GN_Y"
ytlist = "https://www.youtube.com/playlist?list=PLWIVj90xdDE-0js8Ui1Z__uPk3Vt4BYy6"
out_dir = "/content/ytd"
plist = Playlist(ytlist)
for yturl in plist.video_urls:
  download_ytvideo_and_sub(yturl, out_dir)
```
